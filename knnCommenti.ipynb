```python
#1 importo le librerie necessarie

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
# Carica il dataset Iris
iris = load_iris()
# Separa i dati (caratteristiche) e i target (classi)
x = iris.data
y = iris.target
# Crea una nuova figura di dimensioni 10x10"
plt.figure(figsize=(10,10))
# Crea un grafico a dispersione delle caratteristiche
# Utilizza il primo e il secondo attributo di ogni fiore
# Colora i punti in base alla classe del fiore (y)
plt.scatter(x[:,0], x[:,1], c=y, marker="+", s=100, edgecolors="yellow")
# Mostra grafico
plt.show()
```

```python
# Divide il dataset in set di addestramento e di test
# Random_state garantisce che la divisione sia replicabile
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state = 0)
# Stampa i dati di addestramento e di test
print("X_train", X_train)
print("X_test", X_test)
```

```python
# 3. Scelta k
n_neighbours = 5
```

```python
#4. Apprendimento
# Crea un'istanza del classificatore K-Nearest Neighbors con 5 vicini
knn = KNeighborsClassifier(n_neighbors= 5)
# Addestra il modello K-Nearest Neighbors usando i dati di addestramento
knn.fit(X_train, y_train)
```

```python
# Utilizza il modello addestrato per fare previsioni sui dati di test e stamparne le previsioni
y_pred = knn.predict(X_test)
print(y_pred)
```

```python

#calcola l'accuratezza delle previsioni
accuracy = accuracy_score(y_pred, y_test)
#calcola l'accuratezza delle previsioni utilizzando il metodo "micro"
precision = precision_score(y_pred,y_test, average="micro")
#stampa accuratezza e precisione
print(accuracy)
print(precision)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
# Calcola la matrice di confusione
cm = confusion_matrix(y_pred, y_test)
# Crea una visualizzazione della matrice di confusione
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
#Mostra la visualizzazione della matrice di confusione
disp.plot()
```
